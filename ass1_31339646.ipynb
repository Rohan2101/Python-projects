```python
#Rohan Singh
#31339646
#17-08-2022
#25-08-2022
"""
Description of the program:
A Python program with only a text interface that will allow a user (player) to play a dice-rolling game
The aim of the Dice Rolling Game is for a player to roll 2-6 dice multiple times in each game.
Victory is achieved when the median value is the same as the mean value in a result list of
the dice rolls for all rounds of the game. The game starts by letting the player choose a number of dice to play with 
and also choose the type of dice that they want to use. The player rolls the dice multiple times and based on
the results, either wins or loses.
The main menu function will display the main menu when the program is started and has three options.
The game menu function will display the game menu, with a possible 5 options.
The roll dice function rolls the dice given the player’s input for the number and type of dice. The
function returns a tuple of integers each of which represents the dice number. 
The check win/lose function checks the win/lose status for the dice results of the player rolled. Once the function
gets a result list of the dice rolls for all rounds of the game, it finds the middle value (median) and mean value
(average) of this list. If the middle value is the same as the mean value, the function will return the “win” string,
otherwise, return the “lose” string. If the number of rolls for any particular game is over 10
times, return the “lose” string.
The games history function displays all the games’ history.
In the main function the main control logic is implemented in this function.
"""





import random                       # importing random for getting a random dice toss
dice_result_list = []               # initializing empty lists 
games_result_list = []              # initializing empty lists
games_result_list1 = []             # initializing empty lists
games_result_list2 = []             # initializing empty lists
dice_no1 = 2                        # initializing default values for variables
ty_dice1 = 1                        # initializing default values for variables
def mainfunc():                       # main function
    def no_dice():                    # number of dice function
        dice_no = input("Enter the number of dice from 2 to 6:\n")
        if (dice_no not in ['2','3','4','5','6']):
            print("Incorrect number")                   #handling the errors in input
            no_dice()
            return 2
        else: 
            return dice_no
    
    def ty_dicef():                                      # type of dice function
        type_dice = input("Input the type of dice:\n")
        if (type_dice not in ['6','8','9']):
            print("wrong input")                          #handling the errors in input
            ty_dicef()
            return 6
        else:
            return type_dice
        
    
    def rolldice(dice_no=3, ty_dice=6):             # roll dice funtion taking 2 parameters dice number and type of dice
        rolled =[]                                          # initializing empty list
        if ty_dice == 6:
            for no in range(1,(dice_no+1)):
                dice = random.randint(1,6)                    #getting a random number from 1 to 6
                rolled.append(dice)
            t = tuple(rolled)
            return t
            
        elif ty_dice == 8:
            for no in range(1,(dice_no+1)):
                dice = random.choice([2,2,3,3,4,8])           #getting a random number from 2,2,3,3,4,8 
                rolled.append(dice)
            t = tuple(rolled)
            return t
            
        elif ty_dice == 9:
            for no in range(1,(dice_no+1)):
                dice = random.choice([1,1,1,1,5,9])            #getting a random number from 1,1,1,1,5,9
                rolled.append(dice)
            t = tuple(rolled)
            return t  
        
    def winlose_func(dice_result_list):                      # win loss function taking a dice result list parameter
        if len(dice_result_list) > 10:                       # if more than 10 rolls then loss
            return "LOSS"
         #iterate through the sublist using List comprehension to find median
        lst = [element for innerList in dice_result_list for element in innerList]
        n = len(lst)
        lst.sort()                                                  #sorting the list
        if n % 2 ==0:                                               #finding median
            median1 = lst[n//2]
            median2 = lst[n//2 - 1]
            median = (median1 + median2)/2
        else:
            median = lst[n//2]
        median = int(median)
        sm = sum(lst)                                            #finding the mean
        mean = sm/n
        mean = int(mean)
        if median == mean:                                     #checking if median is equal to mean
            return "WIN!"
        else:
            return "LOSS"
        
    def game_hist(x):                                          #game history function
        x.sort()
        for item in x:
            print(f'Dice rolling times: {item[0]}')
            if item[1] == []:
                item[1] = [2]
            print(f'Number of dice: {item[1]}')
            print(f'Result: {item[2]}')
            
    
    def game_menu():                                                            #game menu function 
        print("""               ----------Game Menu--------""")
        print()
        global dice_result_list                                           #defining global variables
        global games_result_list
        global games_result_list1
        global games_result_list2
        global ty_dice1
        global dice_no1
        if ty_dice1 == 1:                                                #checking if type of dice has been input yet
            ty_dice1 = int(ty_dicef())

        choice = input("""
                 A: Press 1 for choosing the number of dice
                 B: Press 2 for rolling the dice
                 C: Press 3 for Checking Win/Lose
                 D: Press 4 to exit the game and return to the main menu
                 
                 Please enter your choice:  """)                                #inputs the game menu choice
        if choice == '1':
            dice_no1 = int(no_dice())
            games_result_list2.append(dice_no1)
            game_menu()
        elif choice == '2':
            dice_result = rolldice(dice_no1, ty_dice1)
            dice_result_list.append(dice_result)
            print(dice_result)
            dice_str = dice_str_func(dice_result)
            print(f'\n{dice_str}')               
            game_menu()
        elif choice == '3':
            if dice_result_list == []:
                print("You haven't rolled dices yet")
                game_menu()
            print(winlose_func(dice_result_list))
            games_result_list1.append(len(dice_result_list))
            games_result_list1.append(games_result_list2)
            games_result_list1.append(winlose_func(dice_result_list))
            games_result_list.append(games_result_list1)
            dice_result_list = []
            games_result_list1 = []
            games_result_list2 = []
            ty_dice1 = 1
            game_menu()
        elif choice == '4':
            main_menu()
        else:
            print("""
                    you input a wrong variable, Please try again!""")             #handling the errors in input
            game_menu()
        
    def dice_str_func(dice_result):                            #function to visualise the string dice
        dice_row = dicer(dice_result)
        heading = " visualised string dice ".upper()
        vdice = "\n".join([heading] + dice_row)
        return vdice

    def dicer(dice_result):                                     # function to concat dices
        dices = []                                              # intializing empty list
        for v in dice_result:
            dices.append(dice[v])
        dice_row = []                                           # intializing empty list
        for r in range(len(dice[1])):
            row = []
            for d in dices:
                row.append(d[r])
            rows = " ".join(row)
            dice_row.append(rows)
        return dice_row        
    dice = {                                                   # dict with dice strings
        1: (
            "---------",
            "|       |",
            "|   o   |",
            "|       |",
            "---------",
        ),
        2: (
            "---------",
            "|       |",
            "| o   o |",
            "|       |",
            "---------",
        ),
        3: (
            "---------",
            "| o     |",
            "|   o   |",
            "|     o |",
            "---------",
        ),
        4: (
            "---------",
            "| o   o |",
            "|       |",
            "| o   o |",
            "---------",
        ),
        5: (
            "---------",
            "| o   o |",
            "|   o   |",
            "| o   o |",
            "---------",
        ),
        6: (
            "---------",
            "| o   o |",
            "| o   o |",
            "| o   o |",
            "---------",
        ),
        8: (
            "---------",
            "| o o o |",
            "| o   o |",
            "| o o o |",
            "---------",
        ),
        9: (
            "---------",
            "| o o o |",
            "| o o o |",
            "| o o o |",
            "---------"
        ),
    }

      
    
    def main_menu():
        print("""               -----------Main Menu------------ """)           # main menu function
        main_menu_var = input('\nPlease press 1 for the main game menu \npress 2 for your game history \nPress 3 if you want to exit \n')
        if main_menu_var == '1':
            game_menu()
        elif main_menu_var == '2':
            game_hist(games_result_list)
            main_menu()
        elif main_menu_var == '3':
            exit()
        else:
            print('you input a wrong variable')    
            main_menu()


    main_menu()
mainfunc()    
```
